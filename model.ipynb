```python
import ast
```

```python
def convert(text):
    L = []
    # Check if the text is a string before evaluating
    if isinstance(text, str):
        try:
            data = ast.literal_eval(text)
            for i in data:
                L.append(i['name'])
        except (ValueError, SyntaxError) as e:
            print(f"Error parsing: {text} -> {e}")
    else:
        # If it's already a list/tuple, handle it directly
        for i in text:
            L.append(i['name'] if isinstance(i, dict) and 'name' in i else i)
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    if pd.isna(text):  # Handle NaN values
        return L
    try:
        for i in ast.literal_eval(text):
            if len(L) < 3:
                L.append(i['name'])
            else:
                break
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing: {text}\n{e}")
    return L
```

```python

movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    # Check if text is a string before evaluating
    if isinstance(text, str):
        try:
            data = ast.literal_eval(text)
        except (ValueError, SyntaxError) as e:
            print(f"Error parsing: {text} -> {e}")
            data = []
    else:
        # If already a list/tuple, assign directly
        data = text

    # Extract director names
    for i in data:
        if isinstance(i, dict) and i.get('job') == 'Director':
            L.append(i.get('name'))
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python

movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else x)
```

```python

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
print(new['tags'].head())
print(new['tags'].isnull().sum())
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
